```python
df = pd.read_csv('./data/preprocessed_movies_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import category_encoders as ce
df = df.drop(columns=['Languages', 'genres', 'countries_origin', 'filming_locations', 'writers'])

X = df.drop('grossWorldWide', axis=1)
y = df['grossWorldWide']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
c_columns = ['Category', 'directors', 'stars', 'production_companies']

encoder = ce.TargetEncoder(cols=c_columns)
X_train_encoded = encoder.fit_transform(X_train, y_train)
X_test_encoded = encoder.transform(X_test)

print(X_train_encoded.head())
```
